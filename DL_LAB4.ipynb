# Multiple Linear Regression from Scratch

We implement a linear regression model using only NumPy and train it with gradient descent on a salary prediction dataset containing age and experience as features.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("multiple_linear_regression_dataset.csv")

print("First 5 rows:\n", df.head())
print("\nColumn Names:", df.columns)
print("Shape of Dataset:", df.shape)
```

## Feature and Target Separation

We extract age and experience as input features and income as the target variable.

```python
X = df[["age", "experience"]].values
y = df["income"].values
```

## Parameter Initialization

Weights and bias are initialized to zero. We need one weight per input feature so each feature contributes independently to the prediction.

```python
num_features = X.shape[1]
weights = np.zeros(num_features)
bias = 0.0
```

## Forward Pass

The prediction is a linear combination of inputs and weights plus a bias term: $\hat{y} = X \cdot w + b$. No activation function is used since we need a continuous output.

```python
def linear_predict(X, weights, bias):
    return X.dot(weights) + bias
```

## MSE Loss Function

Mean Squared Error penalizes large prediction errors more heavily due to squaring. It also produces smooth gradients which are better suited for optimization.

```python
def mse(actual, predicted):
    return ((predicted - actual) ** 2).mean()
```

## Gradient Computation

The gradients of MSE with respect to weights and bias give us the direction to move parameters in order to reduce the loss.

```python
def get_gradients(X, actual, predicted):
    n = len(actual)
    residual = predicted - actual
    grad_w = (2 / n) * X.T.dot(residual)
    grad_b = (2 / n) * residual.sum()
    return grad_w, grad_b
```

## Parameter Update

Gradient descent moves the parameters in the direction that reduces loss. The learning rate controls the step size.

```python
def update_params(weights, bias, grad_w, grad_b, lr):
    weights = weights - lr * grad_w
    bias    = bias    - lr * grad_b
    return weights, bias
```

## Training

We run gradient descent for 1000 epochs with a small learning rate since income values are large numbers.

```python
learning_rate = 0.0001
n_epochs = 1000

for ep in range(n_epochs):
    y_pred = linear_predict(X, weights, bias)
    current_loss = mse(y, y_pred)
    grad_w, grad_b = get_gradients(X, y, y_pred)
    weights, bias = update_params(weights, bias, grad_w, grad_b, learning_rate)

    if ep % 100 == 0:
        print(f"Epoch {ep}, Loss: {current_loss}")
```

## Results and Prediction

After training, we print the learned weights and use them to predict income for a new candidate with age 50 and 15 years of experience.

```python
print(f"Final Weights (age, experience): {weights}")
print(f"Final Bias: {bias}")

sample = np.array([50, 15])
predicted_salary = sample.dot(weights) + bias
print(f"Predicted income for [50, 15]: {predicted_salary}")
```
